```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("../datasets/heart.csv")
df.head()
```

```python
#check the dataset info
df.info()
```

```python
#checking missing values, there is none
df.isnull().sum()
```

```python
#checking duplicates, there is none
df.duplicated().sum()
```

```python
#checking data type, found object it need encoding because the ML work with numbers
df.dtypes
```

```python
#encoding all data of type object
le = LabelEncoder()
for col in df.select_dtypes(include='object'):
    df[col] = le.fit_transform(df[col])
```

```python
#checking again that its fixed
df.dtypes
```

```python
df.head()
```

```python
#exporting the new clean dataset
df.to_csv("../processed_data/heart_clean.csv", index=False)
print("Cleaned dataset saved successfully.")
```
